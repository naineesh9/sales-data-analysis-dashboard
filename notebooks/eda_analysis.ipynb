# Sales Data Analysis & Dashboard

This notebook cleans retail sales data, performs exploratory data analysis, and produces summary outputs that support a Power BI dashboard.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/sales_data.csv')
df.head()
```

```python
# Basic cleaning
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Month'] = df['Order Date'].dt.to_period('M').astype(str)
df['Profit Margin'] = df['Profit'] / df['Sales']

print(df.info())
print('\nMissing values by column:')
print(df.isna().sum())
```

```python
# KPI summary
summary = {
    'Total Sales': round(df['Sales'].sum(), 2),
    'Total Profit': round(df['Profit'].sum(), 2),
    'Total Orders': int(df['Order ID'].nunique()),
    'Profit Margin': round(df['Profit'].sum() / df['Sales'].sum(), 4)
}
summary
```

```python
# Regional sales
region_sales = df.groupby('Region', as_index=False)['Sales'].sum().sort_values('Sales', ascending=False)
region_sales
```

```python
# Monthly sales trend
monthly_sales = df.groupby('Month', as_index=False)['Sales'].sum()
plt.figure(figsize=(10, 4))
plt.plot(monthly_sales['Month'], monthly_sales['Sales'])
plt.xticks(rotation=45)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
# Top sub-categories by sales
top_subcats = df.groupby('Sub-Category', as_index=False)['Sales'].sum().sort_values('Sales', ascending=False).head(10)
top_subcats
```

## Recommended Power BI visuals

- KPI cards: Total Sales, Total Profit, Total Orders, Profit Margin
- Line chart: Monthly Sales Trend
- Bar chart: Sales by Region
- Bar chart: Top Sub-Categories by Sales
- Filters/Slicers: Region, Category, Segment, Month
